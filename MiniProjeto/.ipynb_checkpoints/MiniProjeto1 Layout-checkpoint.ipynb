```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import stats

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_csv("suiciderate.csv")
```

```python
dados
```

```python
tab = pd.DataFrame()

tab["Região"] = dados.Area
tab["Suicídios"] = dados.Count

tab
```

```python
faixa = np.arange(0,2000,100)
tab.Suicídios.hist(bins=faixa, normed=True, color="darksalmon", figsize = (8, 8))
plt.ylabel("Densidade")
```

```python
tab.Suicídios.mode()
```

```python
tab.Suicídios.median()
```

```python
tab.Suicídios.mean()
```

## A media e maior que a mediana que é maior que a moda. Logo temos uma asimetria positiva
