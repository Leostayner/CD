### Itens 7 a 9 - Resolução:

```python
mu_exp =dados.Suicidio.mean()

# k para a gamma
k_gamma = 1.5

# Estimamos theta
theta_gamma = dados.Suicidio.mean()/k_gamma


x = np.linspace(0, 1800, 100)

fig = plt.figure(figsize=(14,10))


# histograma e densidade teórica
dados.Suicidio.hist(bins=faixa, normed=True, data=tempos, color="darksalmon", figsize = (8, 8))
exponencial = stats.expon.pdf(x, scale=mu_exp)
gamma = stats.gamma.pdf(x, scale=theta_gamma, loc=-5, a=k_gamma)

plt.plot(x, gamma, '-', color='green', label="Gamma") # Plota a gamma teórica
plt.plot(x, exponencial, '-', color='red', label="Exponencial") # Plota a exponencial teórica

plt.legend()
plt.axis([0, 1800, 0, 0.0050])
plt.show()
```

```python
# dados ordenados e quantis teóricos - QQplot

# Calcula a frequência relativa acumulada com os dados vindos dos arquivos
fra = [] 
for i in range(len(dados.Suicidio)):
    acc = (i - 0.5) / len(dados.Suicidio)
    if acc < 0:
        acc = 0
    fra.append(acc) 


fra = pd.Series(data=fra)

# Gera os valores dos quantis teóricos para as distribuições analisadas
q_exp = stats.expon.ppf(fra, scale=mu_exp) 
q_gamma = stats.gamma.ppf(fra, scale=theta_gamma, loc=-5, a=k_gamma)


fig = plt.figure(figsize=(12,8))

plt.title("Q-Q plot: Dados x Exponencial x Gamma")

exp_plot, = plt.plot(dados.Suicidio, q_exp, 'o', color="red", label="Exponencial")
gamma_plot, = plt.plot(dados.Suicidio, q_gamma, 'o', color="green", label = "Gamma")

plt.legend(handles=[exp_plot, gamma_plot])
plt.plot((min(dados.Suicidio), max(dados.Suicidio)), (min(dados.Suicidio), max(dados.Suicidio)), '-')


for i in range(len(dados.Suicidio)):
    plt.plot((dados.Suicidio[i], dados.Suicidio[i]), (q_exp[i], dados.Suicidio[i]), '-', color='red', alpha=0.3  )
    plt.plot((dados.Suicidio[i], dados.Suicidio[i]), (q_gamma[i], dados.Suicidio[i]), '-', color='green', alpha=0.3  )

```

```python
# fra e FDA
cdf_exponencial = stats.expon.cdf(dados.Suicidio, scale=mu_exp) #FDA
cdf_gamma = stats.gamma.cdf(dados.Suicidio, scale=theta_gamma, loc=0, a=k_gamma)


fig = plt.figure(figsize=(12,8))

plt.title("Cumulativa: Dados x Exponencial x Gamma")

dados_fra_plot, = plt.plot(dados.Suicidio, fra, '-', color='black', label="F.R.A dos dados")
cdf_exp_plot,   = plt.plot(dados.Suicidio, cdf_exponencial, '-', color='red', label = "Cumulativa da exponencial")
cdf_gamma_plot, = plt.plot(dados.Suicidio, cdf_gamma, '-', color='green', label = "Cumulativa da distribuição Gamma")

plt.legend(handles=[dados_fra_plot, cdf_exp_plot, cdf_gamma_plot])


for i in range(len(dados.Suicidio)):
    plt.plot((dados.Suicidio[i], dados.Suicidio[i]), (fra[i], cdf_exponencial[i]), ':', color="red", alpha=0.3)
    #Omitido para não poluir muito o gráfico
    plt.plot((dados.Suicidio[i], dados.Suicidio[i]), (fra[i], cdf_gamma[i]), '--', color="green", alpha=0.3)
    
```
